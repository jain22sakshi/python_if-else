```python
#Question 6
# Write a Python program to find the largest among three numbers entered by the user.

#Answer:
a = float(input("Enter the First Number : "))                   # float changes the type of input into float(which takes decimals)
b = float(input("Enter the Second Number : "))
c = float(input("Enter the Third Number : "))

if a > b > c or a > c > b :
    print("The Largest Number among all three is : ",a)
    
elif b > a > c or b > c > a :
    print("The Largest Number among all three is : ",b)         #The elif function is short for else if.
elif c > a > b or c > b > a:
    print("The Largest Number among all three is : ",c)
else :
    print("All three Numbers are equal.")
    

 
                
```

```python
#Question 7
#Create a simple calculator program that takes two numbers and an operator (+, -, *, /) as input and performs the corresponding operation.

#Answer:
a = float(input("Enter the First Number : "))
b = float(input("Enter the Second Number : "))
operation = input("Define the mathematical operation(+ ,-,/,*): ")

if operation == "+":
          print("The Operation provided is summation : ",a+b)
        
elif operation == '-':
          print("The operation provided is difference : ",a-b)
        
elif operation == "*":
          print("The Operation provided is Multiplication : ",a*b)
        
elif operation == "/" :
    if b !=0 :
        print("The Operation provided is Division : ",a/b)
    else :
        print("The Division cannot be performed")
        
else :
    print("Invalid Operation")
          
```

```python
#Question 10
# Aceept three sides of triangle and check whether the triangle is equvilateral, isosceles , or scalene 
#(Equvilateral = all three sides are equal)  , (isosceles = any two sides are equal)  ,  (Scalene = all the three sides are unequal)

#Answer
a = float(input("Enter the value for first side: "))
b = float(input("Enter the value for second side: "))
c = float(input("Enter the value for third side: "))

if a == b == c :
    print("This is an Equvilateral Triangle")
elif a==b and a!=c or a==c and a!=b:                          # Here we have to give two elif statement as in forst statement it checking the equality of a with b and c and in the second elif statement it is checking the equality of b with c.
    print("This is an Isosceles Triangle")
elif b==c and b!=a :
     print("This is an Isosceles Triangle")
else:
    print("This is an Scalene Triangle")
```

```python
## Question 13
#Develop a program that calculates the roots of a quadratic equation (ax^2 + bx + c = 0) and handles cases where the discriminant is positive, zero, or negative.
#If D is positive, there are two distinct real roots.
#If D is zero, there is one real root (repeated).
#If D is negative, there are complex roots.

#Answer
a = float(input("Enter the coefficient of a: "))
b = float(input("Enter the coefficient of b: "))
c = float(input("Enter the coefficient of c: "))
D = b**2-4*a*c
if D>0 :
    root1 = (-b + (D**0.5))/(2*a)
    root2 = (-b - (D**0.5))/(2*a)
    print("The roots are Real and Distinct")
    print("Root1 = ",root1)
    print("Root2 = ",root2)
elif D==0:
    root = (-b)/(2*a)
    print("The root is Real and Repeated.")
    print("Root = ",root)
else:
    real_part = (-b)/(2*a)
    imaginary_part = (-(D)**0.5)/(2*a)
    print("The Roots are Complex.")
    print("Root 1 =", real_part, "+", imaginary_part, "i")
    print("Root 2 =", real_part, "-", imaginary_part, "i")
```
